### Montando um Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.

```python
df_train = train['Mensagem'].to_frame()
df_train['Target'] = train['Target']

df_test = test['Mensagem'].to_frame()
df_test['Target'] = test['Target']




```

```python
caracteres = [':',',','.','(',')', '?', '!', '"']
i = 0
for msg in df_train.Mensagem:
    for caractere in caracteres:
        if caractere in msg:
            msg = msg.replace(caractere, '')
            msg = msg.lower()
    df_train.loc[i, 'Mensagem'] = msg 
    i += 1
    
    
j = 0
for msg in df_test.Mensagem:
    for caractere in caracteres:
        if caractere in msg:
            msg = msg.replace(caractere, '')
            msg = msg.lower()
    df_test.loc[j, 'Mensagem'] = msg 
    j += 1




```

```python
livro_conteudo = ''
df_train_conteudo = df_train.loc[(df_train.Target == 'conte√∫do')]
for msg in df_train_conteudo.Mensagem:
    livro_conteudo +=msg

livro_entrega = ''
df_train_entrega = df_train.loc[(df_train.Target == 'entrega')]
for msg in df_train_entrega.Mensagem:
    livro_entrega += msg


livro_preco = ''
df_train_preco = df_train.loc[(df_train.Target == 'pre√ßo')]
for msg in df_train_preco.Mensagem:
    livro_preco +=msg

livro_geral = livro_entrega + livro_conteudo + livro_preco

livro_preco
```

### Verificando a performance do Classificador

Agora voc√™ deve testar o seu classificador com a base de Testes.

```python
lista_mensagem_treino = pd.Series(df_train.Mensagem)

lista_dos_splits = []
for msg in lista_mensagem_treino:
        for palavra in msg.split():
            lista_dos_splits.append(palavra)


universo = list(set(lista_dos_splits))


print(universo)

print(len(universo))
```

```python
#suaviza√ß√£o de laplace
frequencia_suavizada = dict()
serie_geral = pd.Series(livro_geral.split())
tabela_geral = serie_geral.value_counts(True)
tabela_geral

serie_conteudo = pd.Series(livro_conteudo.split())
tabela_conteudo = serie_conteudo.value_counts(True)
tabela_conteudo

serie_entrega = pd.Series(livro_entrega.split())
tabela_entrega = serie_entrega.value_counts(True)
tabela_entrega

serie_preco = pd.Series(livro_preco.split())
tabela_preco = serie_preco.value_counts(True)
tabela_preco

prob_entrega = len(serie_entrega)/(len(serie_entrega) + len(serie_conteudo) + len(serie_preco))
prob_conteudo = len(serie_conteudo)/(len(serie_entrega) + len(serie_conteudo) + len(serie_preco))
prob_preco = len(serie_preco)/(len(serie_conteudo) + len(serie_entrega) + len(serie_preco))



#suavizador
dicionario_de_prob_palavra_tal_entrega = dict()
for frase in df_test['Mensagem']:
    for palavra in frase.split():
        if palavra in tabela_entrega:
            if palavra not in dicionario_de_prob_palavra_tal_entrega:
                dicionario_de_prob_palavra_tal_entrega[palavra] = (tabela_entrega[palavra]+1)/(len(livro_geral.split()) + len(universo))
            else:
                dicionario_de_prob_palavra_tal_entrega[palavra] += (tabela_entrega[palavra]+1)/(len(livro_geral.split()) + len(universo))
        else:
            if palavra not in dicionario_de_prob_palavra_tal_entrega:
                dicionario_de_prob_palavra_tal_entrega[palavra] = 1/(len(livro_geral.split()) + len(universo))
            else:
                dicionario_de_prob_palavra_tal_entrega[palavra] += 1/(len(livro_geral.split()) + len(universo))


dicionario_de_prob_palavra_tal_conteudo = dict()
for frase in df_test['Mensagem']:
    for palavra in frase.split():
        if palavra in tabela_conteudo:
            if palavra not in dicionario_de_prob_palavra_tal_conteudo:
                dicionario_de_prob_palavra_tal_conteudo[palavra] = (tabela_conteudo[palavra]+1)/(len(livro_geral.split()) + len(universo))
            else:
                 dicionario_de_prob_palavra_tal_conteudo[palavra] += (tabela_conteudo[palavra]+1)/(len(livro_geral.split()) + len(universo))
        else:
            if palavra not in dicionario_de_prob_palavra_tal_conteudo:
                dicionario_de_prob_palavra_tal_conteudo[palavra] = 1/(len(livro_geral.split()) + len(universo))
            else:
                dicionario_de_prob_palavra_tal_conteudo[palavra] += 1/(len(livro_geral.split()) + len(universo))



dicionario_de_prob_palavra_tal_preco = dict()
for frase in df_test['Mensagem']:
    for palavra in frase.split():
        if palavra in tabela_preco:
            if palavra not in dicionario_de_prob_palavra_tal_preco:
                dicionario_de_prob_palavra_tal_preco[palavra] = (tabela_preco[palavra]+1)/(len(livro_geral.split()) + len(universo))
            else:
                 dicionario_de_prob_palavra_tal_preco[palavra] += (tabela_preco[palavra]+1)/(len(livro_geral.split()) + len(universo))
        else:
            if palavra not in dicionario_de_prob_palavra_tal_preco:
                dicionario_de_prob_palavra_tal_preco[palavra] = 1/(len(livro_geral.split()) + len(universo))
            else:
                dicionario_de_prob_palavra_tal_preco[palavra] += 1/(len(livro_geral.split()) + len(universo))





#classificador
df_resultado = {}
for i,frase in enumerate(df_test['Mensagem']):
    prob_frase_dado_entrega = 1
    prob_frase_dado_conteudo = 1
    prob_frase_dado_preco = 1
    prob_entrega_dado_frase = 0
    prob_conteudo_dado_frase = 0
    prob_preco_dado_frase = 0
    
    for palavra in frase.split():
        n = df_test.iloc[i]
        if n['Target'] == 'entrega':
            prob_frase_dado_entrega *= dicionario_de_prob_palavra_tal_entrega[palavra]
            prob_entrega_dado_frase = prob_frase_dado_entrega*prob_entrega
        elif n['Target'] == 'conte√∫do':
            prob_frase_dado_conteudo *= dicionario_de_prob_palavra_tal_conteudo[palavra]
            prob_conteudo_dado_frase = prob_frase_dado_conteudo*prob_conteudo

        elif n['Target'] == 'pre√ßo':
            prob_frase_dado_preco *= dicionario_de_prob_palavra_tal_preco[palavra]
            prob_preco_dado_frase = prob_frase_dado_preco*prob_preco
    

    comparador_probabilidades = [prob_entrega_dado_frase, prob_conteudo_dado_frase, prob_preco_dado_frase]
        

    if max(comparador_probabilidades) == prob_conteudo_dado_frase:
        df_resultado[i] = 'conte√∫do'
    elif max(comparador_probabilidades) == prob_entrega_dado_frase:
        df_resultado[i] = 'entrega'
    elif max(comparador_probabilidades) == prob_preco_dado_frase:
        df_resultado[i] == 'pre√ßo'
    






    






         
```
