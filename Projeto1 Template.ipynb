### Montando um Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.

```python
df_train = train['Mensagem'].to_frame()
df_train['Target'] = train['Target']

df_test = test['Mensagem'].to_frame()
df_test['Target'] = test['Target']


```

```python
caracteres = [':',',','.','(',')', '?', '!', '"']
i = 0
for msg in df_train.Mensagem:
    for caractere in caracteres:
        if caractere in msg:
            msg = msg.replace(caractere, '')
            msg = msg.lower()
    df_train.loc[i, 'Mensagem'] = msg 
    i += 1
    
    
j = 0
for msg in df_test.Mensagem:
    for caractere in caracteres:
        if caractere in msg:
            msg = msg.replace(caractere, '')
            msg = msg.lower()
    df_test.loc[j, 'Mensagem'] = msg 
    j += 1




```

```python
df_train.head(5)

```

```python
df_test.head(5)

```

```python
livro_conteudo = ''
df_train_conteudo = df_train.loc[(df_train.Target == 'conteúdo')]
for msg in df_train_conteudo.Mensagem:
    livro_conteudo +=msg

livro_entrega = ''
df_train_entrega = df_train.loc[(df_train.Target == 'entrega')]
for msg in df_train_entrega.Mensagem:
    livro_entrega += msg

livro_geral = livro_entrega + livro_conteudo

```

___
### Verificando a performance do Classificador

Agora você deve testar o seu classificador com a base de Testes.

```python
lista_mensagem_teste = pd.Series(df_test.Mensagem)

lista_dos_splits = []
for msg in lista_mensagem_teste:
        for palavra in msg.split():
            lista_dos_splits.append(palavra)


universo = list(set(lista_dos_splits))


print(universo)

print(len(universo))
```

```python
#suavização de laplace
frequencia_suavizada = dict()
serie_geral = pd.Series(livro_geral.split())
tabela_geral = serie_geral.value_counts(True)
tabela_geral

for palavra in livro_geral.split():
    D = len(livro_geral.split()) + len(universo)
    N = tabela_geral[palavra] + 1
    if palavra not in frequencia_suavizada:
        frequencia_suavizada[palavra] = N/D

    else:
        frequencia_suavizada[palavra] +=N/D


         
```
