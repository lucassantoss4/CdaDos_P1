### Montando um Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.

```python
df_train = train['Mensagem'].to_frame()
df_train['Target'] = train['Target']

df_test = test['Mensagem'].to_frame()
df_test['Target'] = test['Target']




```

```python
caracteres = [':',',','.','(',')', '?', '!', '"']
i = 0
for msg in df_train.Mensagem:
    for caractere in caracteres:
        if caractere in msg:
            msg = msg.replace(caractere, '')
            msg = msg.lower()
    df_train.loc[i, 'Mensagem'] = msg 
    i += 1
    
    
j = 0
for msg in df_test.Mensagem:
    for caractere in caracteres:
        if caractere in msg:
            msg = msg.replace(caractere, '')
            msg = msg.lower()
    df_test.loc[j, 'Mensagem'] = msg 
    j += 1




```

### Verificando a performance do Classificador

Agora você deve testar o seu classificador com a base de Testes.

```python
lista_mensagem_treino = pd.Series(df_train.Mensagem)


lista_dos_splits = []
for msg in lista_mensagem_treino:
        for palavra in msg.split():
            if palavra not in lista_dos_splits:
                lista_dos_splits.append(palavra)


universo = lista_dos_splits
# print(len(universo))
```

```python
lista_probs = []

for palavra in universo:
    i = 0
    for msg in lista_mensagem_treino:
        if palavra in msg:
            i += 1
    prob = i/len(lista_mensagem_treino)
    lista_probs.append(prob)
```

```python
lista_conteudo_treino = df_train.loc[(df_train.Target == 'conteúdo')]

lista_dos_splits_conteudo = []
for msg in lista_conteudo_treino['Mensagem']:
    for palavra in msg.split():
        if palavra not in lista_dos_splits:
            lista_dos_splits_conteudo.append(palavra)


universo_conteudo = lista_dos_splits_conteudo

lista_probs_conteudo = []

for palavra in universo_conteudo:
    i = 0
    for msg in lista_conteudo_treino['Mensagem']:
        if palavra in msg:
            i += 1
    prob = i/len(lista_conteudo_treino['Mensagem'])
    lista_probs_conteudo.append(prob)
```

```python
lista_entrega_treino = df_train.loc[(df_train.Target == 'entrega')]

lista_dos_splits_entrega = []
for msg in lista_entrega_treino['Mensagem']:
    for palavra in msg.split():
        if palavra not in lista_dos_splits_entrega:
            lista_dos_splits_entrega.append(palavra)


universo_entrega = lista_dos_splits_entrega

lista_probs_entrega = []

for palavra in universo_entrega:
    i = 0
    for msg in lista_entrega_treino['Mensagem']:
        if palavra in msg:
            i += 1
    prob = i/len(lista_entrega_treino['Mensagem'])
    lista_probs_entrega.append(prob)
```

```python
lista_preco_treino = df_train.loc[(df_train.Target == 'preço')]

lista_dos_splits_preco = []
for msg in lista_preco_treino['Mensagem']:
    for palavra in msg.split():
        if palavra not in lista_dos_splits_preco:
            lista_dos_splits_preco.append(palavra)


universo_preco = lista_dos_splits_preco

lista_probs_preco = []

for palavra in universo_preco:
    i = 0
    for msg in lista_preco_treino['Mensagem']:
        if palavra in msg:
            i += 1
    prob = i/len(lista_preco_treino['Mensagem'])
    lista_probs_preco.append(prob)
```

```python
probCont = len(lista_conteudo_treino['Mensagem'])/len(lista_mensagem_treino)
probEnt = len(lista_entrega_treino['Mensagem'])/len(lista_mensagem_treino)
probPreco = len(lista_preco_treino['Mensagem'])/len(lista_mensagem_treino)
```

```python
classificado = []

for msg in lista_mensagem_treino:
    palavras = msg.split()
    prob_msgCont = 1
    prob_msgEnt = 1
    prob_msgPreco = 1
    for palavra in palavras:
        if palavra in universo_conteudo:
            locContPalavra = universo_conteudo.index(palavra)
            locPalavra = universo.index(palavra)
            prob_msgCont *= (lista_probs_conteudo[locContPalavra])/(lista_probs[locPalavra])
        if palavra in universo_entrega:
            locEntPalavra = universo_entrega.index(palavra)
            locPalavra = universo.index(palavra)
            prob_msgEnt *= (lista_probs_entrega[locEntPalavra])/(lista_probs[locPalavra])
        if palavra in universo_preco:
            locPrecoPalavra = universo_preco.index(palavra)
            locPalavra = universo.index(palavra)
            prob_msgPreco *= (lista_probs_preco[locPrecoPalavra])/(lista_probs[locPalavra])

    prob_msgCont *= probCont
    prob_msgEnt *= probEnt
    prob_msgPreco *= probPreco
    maior = max(prob_msgCont, prob_msgEnt, prob_msgPreco)
    if maior == prob_msgCont:
        classificado.append('conteúdo')
    elif maior == prob_msgEnt:
        classificado.append('entrega')
    else:
        classificado.append('preço')
```

```python
pd.crosstab(train['Target'],classificado,normalize=True)*100
```
